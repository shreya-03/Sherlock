```python
def user_intermessage_delay(users):
    user_imd = dict()
    for user in users.keys():
        #if len(users[user]['t']) >= 6:
        if user not in user_imd.keys():
            user_imd[user] = []
            for i in range(len(users[user])-1):
                user_imd[user].append(users[user][i+1]-users[user][i])
    return user_imd

def cluster_user_timestamps(filename):
    users = {}
    with open(filename,'r') as f:
        lines = f.readlines()
        #lines = lines[201:]
        for line in lines:
            user = str(line.split(',"u":')[1].split(',"e":')[0].replace('"',''))
            #print count
            if user in users.keys():
                users[user].append(int(line.split('"t":')[1].split(',"u":')[0].replace('"','')))
            else:
                users[user] = []
                users[user].append(int(line.split('"t":')[1].split(',"u":')[0].replace('"','')))
    f.close()
    return users
```

```python
def get_data(filename):
    users = cluster_user_timestamps(filename)
    user_imds = user_intermessage_delay(users)
    imds_list = []
    #print len(user_imd.keys())
    for user in user_imds.keys():
        imds_list.extend(user_imds[user])
        
    return imds_list
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from tfidf import SpellChecker,SlangNormalization,format_line
import seaborn as sns
from statistics import mean

def DocumentNormalization(document):
    spell = SpellChecker()
    slang = SlangNormalization()
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    normalized_document = slang.translator(document)
    normalized_document = ' '.join([lemmatizer.lemmatize(spell.correction(token.lower()).decode('utf-8'),pos="v") for token in normalized_document.split(' ') if token not in stop_words])
    return normalized_document

def cluster_user_msgs(filename):
    users = {}
    with open(filename,'r') as f:
        lines = f.readlines()
        for line in lines:
            user = str(line.split(',"u":')[1].split(',"e":')[0].replace('"',''))
            if user in users.keys():
                msg = format_line(str(line.split(',"m":')[1].split(',"nm":')[0].replace('"','')))
                users[user].append(DocumentNormalization(msg))
            else:
                users[user] = []
                msg = format_line(str(line.split(',"m":')[1].split(',"nm":')[0].replace('"','')))
                users[user].append(DocumentNormalization(msg))
    f.close()
    return users

def plot_similarity(corr,labels,rotation):
    sns.set(font_scale=0.5)
    g = sns.heatmap(corr,xticklabels=labels,yticklabels=labels,vmin=0,vmax=1,cmap="YlOrRd")
    g.set_xticklabels(labels,rotation=rotation)
    g.set_yticklabels(labels,rotation=0)
    g.set_title("jaccard similarity between users")
    plt.show()

#users_similarities = OrderedDict()
user_msgs = cluster_user_msgs('../Data/Real Data/35-40 Viewers#xanttustreamdatabase_new#dip_7777database_chatterscontrolled.txt')
print "Obtained normalized documents"
```

```python
def jaccard_similarity(doc1,doc2):
    intersection = set(doc1).intersection(set(doc2))
    union = set(doc1).union(set(doc2))
    if(len(union)) > 0:
        return len(intersection)/len(union)
    else:
        return 0.0

#msgs = user_msgs[user_msgs.keys()[2]]
#print msgs[6]

#for i in range(len(msgs)):
#    for j in range(len(msgs)):
#        print "index1:"+str(i)+' '+"index2:"+str(j) 
#        print jaccard_similarity(msgs[i],msgs[j])
```

```python
no_users = len(user_msgs.keys())
corr = np.empty([no_users,no_users])
for count_i,user_i in enumerate(user_msgs.keys()):
    for count_j,user_j in enumerate(user_msgs.keys()):
        users_similarities = []
        print "Pair of user:" + str(count_i) + ','+ str(count_j)
        if len(user_msgs[user_i]) > 0 and len(user_msgs[user_j]) > 0:
            for i in range(len(user_msgs[user_i])):
                for j in range(len(user_msgs[user_j])):
                    users_similarities.append(jaccard_similarity(user_msgs[user_i][i],user_msgs[user_j][j]))
            corr[count_i][count_j] = mean(users_similarities)
        else:
            corr[count_i][count_j] = 0.0
#print users_similarities

plot_similarity(corr,user_msgs.keys(),90)
```
